```python
import yfinance as yf
import pandas as pd
import numpy as np
import requests
import io
%pip install beautifulsoup4
from bs4 import BeautifulSoup
```

```python
#get Tickers from S&P 500
url = 'https://en.wikipedia.org/wiki/List_of_S%26P_500_companies'
user_agent = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3'}

request = requests.get(url, headers=user_agent)
soup = BeautifulSoup(request.text, 'html.parser')

table = soup.findAll('table')[0].find_all('tr')

tickers = [row.find_all('td')[0].text.strip() for row in table[1:]]
tickers = [t.replace('.', '-') for t in tickers]
tickers.remove('Q')  # yfinance uses BRK-B instead of BRK.B
tickers.remove('SNDK')
tickers.remove('SOLS')
print(tickers)
```

```python
#Get data from yfinance

start_date = '2020-01-01'
end_date = '2024-10-01'
data = yf.download(tickers, start=start_date, end=end_date, ignore_tz=True)["Close"]


data.to_csv('sp500_data.csv')
print(data.head())
```
